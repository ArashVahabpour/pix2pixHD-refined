```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
import torchvision.transforms as transforms
import torch
from os.path import basename, splitext
```

```python
def binerize(img):
    ret,th1 = cv2.threshold(img,254,255,cv2.THRESH_BINARY)
    return  th1
def histEQ(img):
    binerized_image = binerize(img)
    hist,bins = np.histogram(img[binerized_image/255 == 0].flatten(),256,[0,256])
    output = hist/(img[binerized_image/255 == 0].flatten()).shape[0]
    CSum = output.cumsum(axis=0)
    return CSum[img]*255
```

```python
image_folder_path = '/home/shared/datasets/cars.merged.new/test_B/'
#'/tmp/tmpoy5916ho/test_B/'
```

```python
for image_path in glob.glob(image_folder_path +'*'):
    img = cv2.imread(image_path,0)
    res = histEQ(img)
    cv2.imwrite('/home/shared/datasets/cars.merged.new.eq/test_B/{}.png'.format(splitext(basename(image_path))[0]),res)
```
